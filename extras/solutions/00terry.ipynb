### 1. Create a vector, scalar, matrix and tensor with values of your choosing using `tf.constant()`.


```python
# Creating a scalar
scalar = tf.constant(8)

# Creating a vector
vector = tf.constant([1 , 2 , 3])

# Creating a matrix
matrix = tf.constant([[1 ,2] ,
                      [3, 4]])

# Creating a tensor
tensor = tf.constant(
                      [
                          [
                              [10 , 9 , 8] ,
                              [1 ,2 , 4],
                              [8 , 4 , 2]
                          ]
                          ,
                          [
                              [10 , 9 , 8] ,
                              [1 ,2 , 4],
                              [8 , 4 , 2]
                          ]
                      ]
                     )

scalar , vector , matrix , tensor
```

### 2. Find the **shape**, **rank** and **size** of the tensors you created in 1.

```python
# Finding the shape, rank and size
print(f'The shape of the tensor: {tensor.shape}')
print(f'The rank of the tensor: {tf.size(tensor)}')
print(f'The rank of the tensor: {tf.rank(tensor)}')
```

### 3. Create two tensors containing random values between 0 and 1 with shape `[5, 300]`.


```python
# Set seed
tf.random.set_seed(42)

# Setting up the shape
tensor_1 = tf.random.uniform(shape=(5, 300))
tensor_2 = tf.random.uniform(shape=(5, 300))

tensor_1, tensor_1.shape, tensor_2, tensor_2.shape
```

### 4. Multiply the two tensors you created in 3 using **matrix multiplication**.


```python
# Multiplying
print('Shape before the matrix multiplication')
print(tensor_1.shape , '\n')
print(tensor_2.shape , '\n')

resultant_vector = tf.linalg.matmul(tensor_1 , tf.transpose(tensor_2))  #transpose of the matrix
print(f'After multiplying the resultant vector is: {resultant_vector} \n')

print(f'The resultant vector shape: {resultant_vector.shape}')
```

### 5. Multiply the two tensors you created in 3 using **dot product**.


```python
# Using tensordot
tf.tensordot(tensor_1 , tf.transpose(tensor_2) , axes=1)
```

### 6. Create a tensor with **random values** between 0 and 1 with shape `[224, 224, 3]`.


```python
# Tensor with shape [224 , 224 , 3]
big_tensor = tf.random.uniform(shape = [224 , 224 , 3], minval = 0 , maxval = 1)

big_tensor
```

```python
# Checking the min and max value of our tensor
print(f'Maximum value: {tf.reduce_max(big_tensor)}')
print(f'Minimum value: {tf.reduce_min(big_tensor)}')
```

### 7. Find the **min** and **max values** of the tensor you created in 6.


```python
# Finding the minimum
minimum = tf.math.reduce_min(big_tensor)
print(f'The minimum value is: {minimum}')
```

```python
# Finding the maximum
maximum = tf.math.reduce_max(big_tensor)
print(f'The maximum value is: {maximum}')
```

### 8. Created a tensor with random values of shape `[1, 224, 224, 3]` then **squeeze** it to change the shape to `[224, 224, 3]`.


```python
# Tensor with shape [224 , 224 , 3]
unsqueezed_tensor = tf.random.Generator.from_seed(42)
unsqueezed_tensor = unsqueezed_tensor.normal(shape = (1 , 224 , 224 , 3))

print(f'The shape before squeezing: {unsqueezed_tensor.shape}')
```

```python
# And we are squeezing it
squeezed_tensor = tf.squeeze(unsqueezed_tensor)
print(f'The shape after squeezing: {squeezed_tensor.shape}')
```

### 9. Create a tensor with shape `[10]` using your own choice of values, then find the **index** which has the **maximum value**.


```python
ten_tensor = tf.constant([0 ,1 ,2 ,3, 4, 9, 10, 0.1 , -000.1 , 6])

# Finding the index where value is maximum
max_value = tf.math.argmax(ten_tensor)

print(f'The indices where the value is maximum: {max_value}')
print(f'The maximum value was: {ten_tensor[max_value]}')
```

### 10. **One-hot** encode the tensor you created in 9.

```python
# Throw's an error since I had floating numbers in my tensor, casting to int solved!
tf.cast(ten_tensor , dtype=tf.int32)
```

```python
# One hot encoding the tensor of shape 10
tf.one_hot(tf.cast(ten_tensor , dtype=tf.int32) , depth = 10)
```

## Extra Curriculum

- Try to create a series of tensor functions to calculate your most recent grocery bill (it's okay if you don't use the names of the items, just the price in numerical form).
    - How would you calculate your grocery bill for the month and for the year using tensors?

```python
import tensorflow as tf

# Example: Prices in dollars of your grocery items
prices = tf.constant([10.99, 5.49, 3.29, 12.50, 7.99, 4.25], dtype=tf.float32)

# Quantities purchased for each item
quantities = tf.constant([1, 2, 1, 1, 3, 2], dtype=tf.float32)

print("Prices:", prices)
print("Quantities:", quantities)
```

```python
# Total for each item
item_totals = prices * quantities

# Final grocery bill
grocery_bill = tf.reduce_sum(item_totals)

print("Item totals:", item_totals.numpy())
print("Your grocery bill for this trip is: $", grocery_bill.numpy())
```

```python
monthly_bill = grocery_bill * 4
print("Your estimated monthly grocery bill is: $", monthly_bill.numpy())
```

```python
yearly_bill = monthly_bill * 12
print("Your estimated yearly grocery bill is: $", yearly_bill.numpy())

# Explanation of tf.reduce_sum and tf.reduce_mean:

# tf.reduce_sum:
# This TensorFlow operation computes the sum of elements across the dimensions of a tensor.
# In your grocery bill example, `tf.reduce_sum(item_totals)` was used to sum up
# all the individual item totals (price * quantity) to get the final grocery bill for the trip.
# It effectively collapses the tensor along specified axes (or all axes if not specified) by summing its values.

# tf.reduce_mean:
# This TensorFlow operation computes the mean (average) of elements across the dimensions of a tensor.
# It calculates the sum of all elements and then divides by the number of elements.
# It's particularly useful for finding the average value within a tensor.

import tensorflow as tf

# Example of tf.reduce_mean using your `prices` tensor:
# Let's find the average price of the grocery items from your list.
# The 'prices' tensor is available from a previous cell.
average_item_price = tf.reduce_mean(prices)
print(f"\nAverage price of an item in your list: ${average_item_price.numpy():.2f}")

# For further illustration, let's use a simple tensor to show both:
test_tensor = tf.constant([10, 20, 30, 40, 50])
print(f"Test Tensor: {test_tensor.numpy()}")
print(f"tf.reduce_sum on Test Tensor: {tf.reduce_sum(test_tensor).numpy()}") # Sums all elements: 10+20+30+40+50 = 150
print(f"tf.reduce_mean on Test Tensor: {tf.reduce_mean(test_tensor).numpy():.2f}") # Averages all elements: 150 / 5 = 30.00
```
